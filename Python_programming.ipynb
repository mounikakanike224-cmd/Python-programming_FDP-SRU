**Relational Operators**

```python
num1=10
num2=0
num3=10
str1="good"
str2="afternoon"
print("status=",num1==num2)  # == compares contents
print("status=",num1==num3)
print("status=",num1!=num2)
print("status=",num1!=num3)
print("status=",num1>num2)
print("status=",num2<num3)
print("status=",num1<=num3)
print("status=",num3>=num2)
```

**Logical Operators**

```python
num1=10
num2=0
num3=10
print("status=",num1==num2 and num1==num3)
print("status=",num1==num2 or num1==num3)
print("status=",not (num1>=num3))


```

**Assignment operators(*=, /=, //=, power(**)=)**

```python
num1=10
print("num1=",num1)
num2=num1
print("num2=",num2)
num2=20
print("num2=",num2)
num1+=num2 #num1=10+20
print ("num1=",num1)
num1-=num2 #num1=30-20
print("num1=",num1)
num1*=num2 #num1=num1*num2
print("num1=",num1)
num1/=num2 #num1=num1/num2
print("num1=",num1)
num1%=num2 #num1=num1%num2
print("num1=",num1)
num1//=num2 #num1=num1%num2
print("num1=",num1)
num2**=2 #num2=num2*num2
print("num2=",num2)
```

**Bitwise Operator**

```python
a=10 #1010
b=5  #0101
print ("result=", a&b)
print ("result=", a|b)
print ("result=", ~a) #bitwise not, write binary equivalent of a in 8 bit format, add 1, place negative sign
print ("result=", ~b)
print ("result=", a^b)
```
